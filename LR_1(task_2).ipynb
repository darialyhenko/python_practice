```python
import random
class Warrior:
    def __init__(self, name, strength = 20, health = 100):
        self.name = name
        self.strength = strength
        self.health = health
    def hit(self,strength):
        self.health = self.health - strength
        return self.health
    def __str__(self):
        return f"{self.name}-й воин, сила атаки - {self.strength}, осталось здоровья - {self.health} HP"
    def __del__(self):
        print('%s-й воин отправляется в Вальгаллу\n'  % self.name)
 
unit1 = Warrior('1', 10,100)
unit2 = Warrior('2',20,100)
unit3 = Warrior('3',50,100)
unit_list = [unit1,unit2,unit3]

while True:
    if len(unit_list)==1:
        print('Победил %s' % str(unit_list[0]))
        break
    x = random.choice(unit_list)
    y = random.choice(unit_list)
    if x != y:
        print('%s-й воин бьет %s-го воина' % (y.name, x.name))
        if x.name == '1':
            print('У %s-го воина осталось %s HP \n' % (x.name, unit1.hit(y.strength)))
            if unit1.health <= 0:
                print('%s-й воин погиб от рук %s-го воина\n' % (x.name, y.name))
                unit_list.remove(x)
                del unit1
        if x.name == '2':
            print('У %s-го воина осталось %s HP \n' % (x.name, unit2.hit(y.strength)))
            if unit2.health <= 0:
                print('%s-й воин погиб от рук %s-го воина\n' % (x.name, y.name))
                unit_list.remove(x)
                del unit2
        if x.name == '3':
            print('У %s-го воина осталось %s HP \n' % (x.name, unit3.hit(y.strength)))
            if unit3.health <= 0:
                print('%s-й воин погиб от рук %s-го воина\n' % (x.name, y.name))
                unit_list.remove(x)
                del unit3
```
